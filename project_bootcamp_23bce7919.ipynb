##Importing Modules required


```python
import pandas as pd
import matplotlib.pyplot as mp
```

##Reading the csv file and displaying it

```python
df = pd.read_csv('Global_Education.csv')
df
```

```python
df.info()
```

##Changing the Index of the data frame

```python
df.set_index(df['Countries and areas'],inplace = True)
df
```

## Removing columns Name of countries, latitude and longitude as they are deeemed useless

```python
df.drop(columns = ['Countries and areas','Latitude ','Longitude'],inplace = True)
df
```

##Displaying the countires having least and highest birth rate

```python
print('Countries having least birth rate: ',df[df['Birth_Rate'] == min(df['Birth_Rate'])].index.to_list())
print('Country having highest birth rate: ',df[df['Birth_Rate'] == max(df['Birth_Rate'])].index.to_list())
```

##Adding two new columns for Average completion rate of Primary , Upper and Lower secondary education

```python
df['average completion rate for primary education'] = (df['Completion_Rate_Primary_Female'] + df['Completion_Rate_Primary_Male'])/2
df['average completion rate for lower secondary education'] = (df['Completion_Rate_Lower_Secondary_Male']+df['Completion_Rate_Lower_Secondary_Female'])/2
df['average completion rate for upper secondary education'] = (df['Completion_Rate_Upper_Secondary_Female'] + df['Completion_Rate_Upper_Secondary_Male'])/2
df
```

## Graphs


```python
x = ['Primary','Lower Secondary','Upper Secondary']
y = [df[col].mean() for col in df.columns[-1:-4:-1]]
mp.bar(x,y,color = 'skyblue',width = 0.4)
mp.xlabel('Level of Education')
mp.ylabel('Average Completion Rate')
mp.title('Average Completion Rate Across Various Education Levels ')
```

```python
temp = df[(df['Youth_15_24_Literacy_Rate_Male']> 0 )& df['Youth_15_24_Literacy_Rate_Female'] > 0]
x1 = temp['Youth_15_24_Literacy_Rate_Male']
x2 = temp['Youth_15_24_Literacy_Rate_Female']
y = temp['Unemployment_Rate']
mp.scatter(x1,y,label = 'Men',alpha = 0.5)
mp.scatter(x2,y,label = 'Women',alpha = 0.5)
mp.xlabel('Literacy Rate')
mp.ylabel('Unemployment Rate')
mp.legend()
mp.title('Literacy Rate vs Unemployment Rate of various countries according to gender')
```

```python
c_m = ['Grade_2_3_Proficiency_Math','Primary_End_Proficiency_Math','Lower_Secondary_End_Proficiency_Math']
c_r = ['Grade_2_3_Proficiency_Reading','Primary_End_Proficiency_Reading','Lower_Secondary_End_Proficiency_Reading']
labels = ['Grade 2-3','Primary End','Lower Secondary End']
width  = 0.2
x = [x for x in range(len(labels))]
ym = [df[x].mean() for x in c_m]
yr = [df[x].mean() for x in c_r]
mp.bar([x-(width/2) for x in x],ym,width,label = 'Math' )
mp.bar([x+(width/2) for x in x],yr,width,label = 'Reading')
mp.xticks(x,labels)
mp.ylabel('Average Proficiency')
mp.xlabel('Level of Education')
mp.legend()
mp.title('Reading and Math Proficiency at different levels of education')
```
